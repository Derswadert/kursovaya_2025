```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.stats import norm, uniform, expon, chi2

def analyze_file(df):
    
    n = len(df)
    g = round(1 + math.log2(n))
    val = df['Value'].astype('float')
    val = val.reset_index(drop=True)
    print(min(val))
    print(max(val))

    h = (val.max() - val.min()) / g
    bins = [val.min() + i * h for i in range(g + 1)]
    bins[0] -= 1e-3 
    intervals = pd.cut(val, bins)

    abs_freq = intervals.value_counts().sort_index()
    rel_freq = round(abs_freq / n, 3)
    return_df = pd.DataFrame({
        'intervals': abs_freq.index,
        'abs_freq': abs_freq,
        'rel_freq': rel_freq
    }).reset_index(drop=True)

    return_df['empf'] = return_df["rel_freq"].cumsum()

    print("Среднее:", val.mean())
    print("Стандартное отклонение:", val.std(ddof=1))
    print("Дисперсия:", val.var(ddof=1))
    print("Коэффициент вариации:", str(round(float(val.std() / val.mean() * 100), 2)) + '%')

    return_df['low_interval'] = return_df['intervals'].apply(lambda x: x.left)
    return_df['center_interval'] = return_df['intervals'].apply(lambda x: x.mid)
    return_df['up_interval'] = return_df['intervals'].apply(lambda x: x.right)

    plt.step(return_df['up_interval'], return_df['empf'], where='post')
    plt.scatter(return_df['up_interval'], return_df['empf'], color='blue')
    plt.xticks(rotation=45)
    plt.locator_params(axis='x', nbins=20)
    plt.title("Эмпирическая функция распределения показателя")
    plt.xlabel("Значение показателя (Value)")
    plt.ylabel("Относительная частота")
    plt.grid(True)
    plt.show()

    left = [interval.left for interval in return_df['intervals']]
    right = [interval.right for interval in return_df['intervals']]
    widths = [r - l for l, r in zip(left, right)]
    freq = return_df['rel_freq']
    
    plt.figure(figsize=(8, 6))
    plt.bar(left, freq, width=widths, align='edge', edgecolor='black')
    plt.xlabel("Значение показателя (Value)")
    plt.ylabel("Относительная частота")
    plt.title("Гистограмма распределения показателя")
    plt.xticks(rotation=45)
    plt.locator_params(axis='x', nbins=20)
    plt.show()

    return return_df
```

```python
from scipy import stats

def analyze_lognormal(col):

    y = np.log(col)
    n = len(col)

    mu_log = np.mean(y)
    s2_log_unbiased = np.var(y, ddof=1)

    mu_mle = mu_log
    sigma2_mle = np.mean((y - mu_mle)**2)

    mu_mm = mu_log
    sigma2_mm = sigma2_mle

    sigma2_bayes = ((n - 1) * s2_log_unbiased) / (n - 3)
    bayes_note = None
    mu_bayes = mu_log

    mean_mle = np.exp(mu_mle + sigma2_mle / 2)
    var_mle = (np.exp(sigma2_mle) - 1) * np.exp(2 * mu_mle + sigma2_mle)

    mean_bayes = np.exp(mu_bayes + sigma2_bayes / 2) if not np.isnan(sigma2_bayes) else np.nan
    var_bayes = ((np.exp(sigma2_bayes) - 1) * np.exp(2 * mu_bayes + sigma2_bayes)
                 if not np.isnan(sigma2_bayes) else np.nan)


    print(f"Среднее log(X): {mu_log:.4f}")
    print()
    print("Оценки параметров логнормального распределения:")
    print(f"  MLE:       mu = {mu_mle:.4f}, sigma² = {sigma2_mle:.4f}")
    print(f"  Моментов:  mu = {mu_mm:.4f}, sigma² = {sigma2_mm:.4f}")
    print(f"  Несмещённая дисперсия log(X): {s2_log_unbiased:.4f}")
    if bayes_note:
        print(bayes_note)
    else:
        print(f"  Байес (Jeffreys prior): mu = {mu_bayes:.4f}, sigma²(post.mean) = {sigma2_bayes:.4f}")
    print()
    print("Ожидание и дисперсия X:")
    print(f"  E[X]_MLE   = {mean_mle:.4f}")
    print(f"  Var[X]_MLE = {var_mle:.4f}")
    if not np.isnan(mean_bayes):
        print(f"  E[X]_Bayes = {mean_bayes:.4f}")
        print(f"  Var[X]_Bayes = {var_bayes:.4f}")

    return {
        "n": n,
        "mu_mle": mu_mle,
        "sigma2_mle": sigma2_mle,
        "mu_mm": mu_mm,
        "sigma2_mm": sigma2_mm,
        "sigma2_log_unbiased": s2_log_unbiased,
        "mu_bayes": mu_bayes,
        "sigma2_bayes": sigma2_bayes,
        "E_X_mle": mean_mle,
        "Var_X_mle": var_mle,
        "E_X_bayes": mean_bayes,
        "Var_X_bayes": var_bayes
    }
```

```python
from scipy.stats import t
def file_evaluation(df):
    n = len(df)

    val = df['Value'].astype('float')
    log_val = np.log(val)
    
    print("Медиана:",np.median(val))
    print("Среднее:", val.mean())
    print("Стандартное отклонение:", val.std(ddof=1))
    print("Дисперсия:", val.var(ddof=1))
    print("Математическое ожидание:", val.mean())
    
    xs = np.linspace(val.min(), val.max(), n*3)
    ys = gaussian_kde(val)
    print("Мода: ",xs[np.argmax(ys)])

    sd = val.std(ddof=1) 
    mean = val.mean()
    t_q = t.ppf(1 - 0.95/2, df=n-1)
    se = sd / np.sqrt(n)
    ci_low = mean - t_q * se
    ci_high = mean + t_q * se
    print(f"Доверительный интервал: [{ci_low}, {ci_high}]")
    
```

```python
from scipy.stats import lognorm, rayleigh, chi2

def _merge(obs, exp, threshold=5):
    obs = list(map(float, obs))
    exp = list(map(float, exp))
    while True:
        idx = next((i for i, e in enumerate(exp) if e < threshold), None)
        if idx is None or len(exp) == 1:
            break
        if idx == 0:
            l, r = 0, 1
        else:
            l, r = idx - 1, idx
        obs[l] += obs[r]
        exp[l] += exp[r]
        del obs[r]
        del exp[r]
    return np.array(obs), np.array(exp)

def run_chi2(df, col, bins=None):

    val = df[col].values
    counts, edges = np.histogram(val, bins=8)
    low, up = edges[:-1], edges[1:]
    ret = pd.DataFrame({"low_interval": low, "up_interval": up, "abs_freq": counts})
    n = len(val)

    mu_log = np.log(val).mean()
    sigma_log = np.log(val).std(ddof=1)
    Ni_norm = (lognorm.cdf(up, s=sigma_log, scale=np.exp(mu_log)) - lognorm.cdf(low, s=sigma_log, scale=np.exp(mu_log))) * n
    ret["Ni_norm"] = Ni_norm

    obs_n, exp_n = _merge(ret["abs_freq"].values, Ni_norm)
    chi2_norm = np.sum((obs_n - exp_n) ** 2 / exp_n)
    v = len(obs_n) - 3
    za = chi2.ppf(0.95, v)
    print(f"χ² (логнормальное): {chi2_norm:.3f}, критическое {za:.3f}")

    sigma_ray = (np.sum(val ** 2) / (2.0 * n)) ** 0.5
    Ni_ray = (rayleigh.cdf(up, scale=sigma_ray) - rayleigh.cdf(low, scale=sigma_ray)) * n
    ret["Ni_rayleigh"] = Ni_ray

    obs_r, exp_r = _merge(ret["abs_freq"].values, Ni_ray)
    chi2_ray = np.sum((obs_r - exp_r) ** 2 / exp_r)
    df_ray = len(obs_r) - 2
    za_ray = chi2.ppf(0.95, df_ray)
    print(f"χ² (Рэлея: {chi2_ray:.3f}, критическое {za_ray:.3f}")

    return ret
```

```python
run_chi2(df_male, "Value")
run_chi2(df_female, "Value")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

def anamorphosis_r2(x, show_plots=True):

    x = np.asarray(x)
    results = {}
    pos = x

    (osm_log, osr_log), (slope_log, intercept_log, r) = st.probplot(np.log(pos), dist='norm', fit=True)
    R2_log = float(r ** 2)
    results['lognorm'] = {
        'R2': R2_log,
        'slope': float(slope_log),
        'intercept': float(intercept_log)
    }

    (osm_ray, osr_ray), (slope_ray, intercept_ray, r) = st.probplot(x, dist='rayleigh', fit=True)
    R2_ray = float(r ** 2)
    results['rayleigh'] = {
        'R2': R2_ray,
        'slope': float(slope_ray),
        'intercept': float(intercept_ray)
    }

    if show_plots:
        fig, axes = plt.subplots(1, 2, figsize=(12, 5))

        ax = axes[0]
        ax.scatter(osm_log, osr_log, s=20)
        ax.plot(osm_log, slope_log * osm_log + intercept_log, linewidth=1)
        ax.set_title(f'ЛогНормальное распределение, R2={R2_log:.4f}')
        ax.set_xlabel('Теоретические квантили')
        ax.set_ylabel('ln(x)')

        ax = axes[1]
        ax.scatter(osm_ray, osr_ray, s=20)
        ax.plot(osm_ray, slope_ray * osm_ray + intercept_ray, linewidth=1)
        ax.set_title(f'Распределение Рэлея, R2={R2_ray:.4f}')
        ax.set_xlabel('Теоретические квантили')
        ax.set_ylabel('x')

        plt.tight_layout()
        plt.show()

    return results
```
